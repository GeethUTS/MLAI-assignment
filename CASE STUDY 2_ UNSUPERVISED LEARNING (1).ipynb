## **b. Try out different clustering models in the Whole sale customer dataset.**

```python
#k means clustering
```



* K-Means Clustering is an Unsupervised Machine Learning algorithm, which groups the unlabeled dataset into different clusters.



*  K-Means Clustering: K-Means is a widely used clustering algorithm. It assigns documents to clusters by minimizing the distance between the document and the cluster's centroid. Each cluster represents a group of similar documents.






```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters = i,n_init=10,init = 'k-means++')
  km.fit(x)
  css = km.inertia_
  wcss.append(css)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No of Clusters')
plt.ylabel('WCSS Score')
plt.grid()
plt.show()
```

```python
km = KMeans(n_clusters = 5,n_init=10,init = 'k-means++')
```

```python
km.fit(x)
```

```python
cluster_labels = km.predict(x)
```

```python
cluster_labels
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
cluster_indices = np.where(cluster_labels==0)
```

```python
cluster_indices
```

```python
data['cluster_labels'] = cluster_labels
```

```python
data.head()
```

```python
colors = ['r','g','b','c','m']

for i in range(max(cluster_labels)+1):
  cluster_data = data[cluster_labels==i]
  plt.scatter(cluster_data['Detergents_Paper'],
              cluster_data['Delicassen'],
              c = colors[i],
              label =f'Cluster{i}')
plt.xlabel('Detergents_Paper')
plt.ylabel('Delicassen')
plt.title('Detergents_Paper vs Delicassen')
plt.legend()
```

```python
colors = ['r','g','b','c','m']

for i in range(max(cluster_labels)+1):
  cluster_data = data[cluster_labels==i]
  plt.scatter(cluster_data['Frozen'],
              cluster_data['Grocery'],
              c = colors[i],
              label =f'Cluster{i}')
plt.xlabel('Frozen')
plt.ylabel('Grocery')
plt.title('Grocery vs Frozen')
plt.legend()
```

```python
#Agglomerative HC
```



* Agglomerative Clustering is a type of hierarchical clustering algorithm. It is an unsupervised machine learning technique that divides the population into several clusters such that data points in the same cluster are more similar and data points in different clusters are dissimilar.




```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings = linkage(x,method='ward')
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings)
plt.show()
```

```python
cluster_labels_hc = fcluster(mergings,250000,criterion = 'distance')
```

```python
cluster_labels_hc
```

```python
colors = ['r','g','b','c','m']

for i in range(max(cluster_labels_hc)+1):
  cluster_data = data[cluster_labels==i]
  plt.scatter(cluster_data['Detergents_Paper'],
              cluster_data['Delicassen'],
              c = colors[i],
              label =f'Cluster{i}')
plt.xlabel('Detergents_Paper')
plt.ylabel('Delicassen')
plt.title('Detergents_Paper vs Delicassen')
plt.legend()
```

```python
#silhouette score
```



*   The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters


* The idea is to calculate the average distance between each object i and all other objects in the same class a(i) Then calculate the distance between each object i and the closest object in all other classes b(i) The silhouette score for object i is defined as: s(i) = (b(i) – a(i)) / max(a(i),b(i))



```python
from sklearn.metrics import silhouette_score
```

```python
sil_score = silhouette_score(x,cluster_labels_hc)
sil_score
```

```python
sil_score = silhouette_score(x,cluster_labels)
sil_score
```

# **Perform DBSCAN Clustering**

```python
#DBSCAN
```



 1. It is a density-based clustering       non-parametric algorithm: given a set of points in some space, it groups together points that are closely packed together, marking as outliers points that lie alone in  -density regions


2.  Clusters are dense regions in the data space, separated by regions of the lower density of points.  



```python
from sklearn.cluster import DBSCAN
```

```python


db = DBSCAN(eps=0.3, min_samples=10).fit(x)
labels = db.labels_

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```

```python
y2 = db.fit_predict(x)
plt.scatter(data['Detergents_Paper'],data['Delicassen'])
plt.xlabel('Detergents_Paper')
plt.ylabel('Delicassen')
plt.title('Detergents_Paper vs Delicassen')
plt.legend()
plt.show()
```
