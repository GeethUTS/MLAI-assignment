```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,flatten,Dense,Dropout
```

```python
(x_train,y_train),(x_test,y_test)=load_data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize(10,10))

for im in range(10);
 plt.subplot(5,4.im+1)
 plt.imshow[x_train[im]]
 plt.show()
```

```python
x_train.shape
```

```python
input_shape=x_train.shape[1:]
input-shape
```

```python
from numpy import unique
num_class=len(unique(y_train))
num_class
```

```python
x_train=x_train.astype('float32')/255.0
```

```python
#create the model
```

```python
model=Sequential()
model.add(Conv2D(32(3,3)),activation='relu',input_shape=input-shape)
model.add(MaxPool2D(2,2))
model.add(flatten())
model.add(Dense(100,activation='relu'))
model.add(Dropout(0,5))
model.add(Dense(num_class,activation='softmax'))
```

```python
model.summary()
```

```python
#compile the model
```

```python
model.compile(optimizer='adam',loss='sparse_categoriacl_crossentropy,metrics=[accuracy]')
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=128,verbose=1)
```

```python
loss,accuracy=model.evaluate(x_test,y_test,verbose=0)
```

```python
image=x_train[32]
import numpy as np
ypred=model.predict(np.arrau([image]))
print('predict=%d'%argmax(ypred))
```
