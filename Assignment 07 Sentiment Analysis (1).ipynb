```python
#import gensim(Gensim is designed to handle large text collections using data streaming and incremental online algorithms, which differentiates it from most other machine learning software packages that target only in-memory processing.)
```

```python
import gensim
data['id,label,tweet_clean'] = data['id,label,tweet'].apply(lambda x:gensim.utils.simple_preprocess(x))
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data['id,label,tweet_clean'],data['id,label,tweet'],test_size =0.3)
```

```python
from gensim.models import Word2Vec
w2v_model = gensim.models.Word2Vec(x_train,min_count = 1)
```

```python
len(w2v_model.wv['with'])
```

```python
w2v_model.wv.most_similar('with')
```

```python
words = w2v_model.wv.index_to_key
```

```python
len(words)
```

```python
x_train_vec = []
x_test_vec = []

for ls in x_train:
  vec = []
  for word in ls:
    if word in words:
      vec.append(w2v_model.wv['word'])
  x_train_vec.append(np.array(vec))
```

```python
for ls in x_test:
  vec = []
  for word in ls:
    if word in words:
      vec.append(w2v_model.wv['word'])
  x_test_vec.append(np.array(vec))
```

```python
x_train_vec_avg = []
x_test_vec_avg = []

def cal_avg_vec(v):
  if v.size:
    return v.mean(axis = 0)
  else:
    np.zeros(100,dtype=float)

for v in x_train_vec:
  x_train_vec_avg.append(cal_avg_vec(v))

for v in x_test_vec:
  x_test_vec_avg.append(cal_avg_vec(v))
```

```python
x_train_vec_avg = np.array(x_train_vec_avg)
x_test_vec_avg = np.array(x_test_vec_avg)
```

```python
#import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(x_train_vec_avg,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test_vec_avg)
```

```python
accuracy_score(y_test,y_pred)
```
