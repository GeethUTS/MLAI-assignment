**DEEP LEARNING**



1.   Deep learning is the subset of machine learning methods based on neural networks with representation learning.

2. The adjective "deep" refers to the use of multiple layers in the network. Methods used can be either supervised, semi-supervised or unsupervised.  



3.   Deep learning is a method in artificial intelligence (AI) that teaches computers to process data in a way that is inspired by the human brain.

4.   Deep learning models can recognize complex patterns in pictures, text, sounds, and other data to produce accurate insights and predictions.

```python
data.drop('emotion_in_tweet_is_directed_at',axis=1,inplace=True)
data.drop('tweet_text',axis=1,inplace=True)
```

```python
import gensim
data['is_there_an_emotion_directed_at_a_brand_or_product_one'] = data['is_there_an_emotion_directed_at_a_brand_or_product'].apply(lambda x:gensim.utils.simple_preprocess(x))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data['is_there_an_emotion_directed_at_a_brand_or_product_one'],data['is_there_an_emotion_directed_at_a_brand_or_product'],test_size=0.2,)
```

```python
y_train.iloc[0]
```

```python
from keras.layers import Dense, LSTM, Embedding
from keras.models import Sequential

model = Sequential()
model.add(Embedding(input_dim=9093, output_dim=5, input_length=9093))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(60))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
model.fit(y_train,y_test,epochs=4)
```

```python
pred_test=model.predict(y_test)>0.5
```
